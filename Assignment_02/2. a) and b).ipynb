### Q. 2. (a)Write python from scratch for 2D Spatial Separable convolution between input image(lena.jpg) and Gausian filter = np.array ([1,4,6,4,1], [4,16,24,16,4], [6,24,36,24,6], [4,16,24,16,4], [1,4,6,4,1])/256

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.tif')

gaussian_kernel = np.array([[1,4,6,4,1], [4,16,24,16,4], [6,24,36,24,6], [4,16,24,16,4], [1,4,6,4,1]])/256
conv_using_cv2 = cv2.filter2D(img, -1, gaussian_kernel, borderType = cv2.BORDER_CONSTANT)

plt.imshow(img)
plt.title("Original Image")
plt.show()

plt.imshow(conv_using_cv2)
plt.title("Filtered Image")
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
