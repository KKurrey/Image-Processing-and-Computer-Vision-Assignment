```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import time
    
img = cv2.imread("GrayImage1.jpg")

cv2.imwrite("GrayImage1.jpg", img)

plt.imshow(img)

plt.title("Grayscale input image")

plt.show()

#Method1- using cv2

t1 = time.time()
ret, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

cv2.imwrite('RoseBinary2.jpg', binary_img)

t2 = time.time()
time_cv2 = t2 - t1
print(time_cv2)

plt.imshow(binary_img)

plt.title("Rose Binary Image")

plt.show()

#Method2 - using def

t3 = time.time()

def gray2binary(img):
    return(cv2.threshold(img, 127, 255, cv2.THRESH_BINARY))

ret, binary_img = gray2binary(img)

t4 = time.time()
time_def = t4 - t3
print(time_def)

plt.imshow(binary_img)

plt.savefig("RoseBinary.jpg")

plt.title("Binary Rose Image")

plt.show()



cv2.waitKey(0)
cv2.destroyAllWindows()

#Comparison of processing speed

x = ['Method-1', 'Method-2']

Processing_times = [time_cv2, time_def]

x_pos = [i for i in range (len(x))]

plt.bar(x_pos, Processing_times, color = 'blue')
plt.xlabel("<-------------Method------------->  ")
plt.ylabel("<-------------Processing Time------------->  ")
plt.title(" Procesing Time versus method ")
plt.xticks(x_pos, x)

plt.show()
```
