### Q. 4. Write python from scratch to convert RGB Image (lena.jpg)  to YCbCr color image using following equation.

Y= 16 + 65.738*R/256 + 129.057*G/256 + 25.064*B/256

Cb = 128-37.945*R/256 - 74.494*G/256 + 112.439*B/256

Cr = 128+112.439*R - 94.154*G/256 - 18.285*B/256

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.tif')
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#plt.imshow(img1)
#plt.title("Original Image")
#plt.plot()

output = img1

for k in range(img1.shape[2]):
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
            if k == 0:
                output[i][j][k] = 16 + (65.738 * img1[i][j][0] / 256) + (129.057 * img1[i][j][1] / 256) + (25.064 * img1[i][j][2] / 256)
            elif k == 1:
                output[i][j][k] = 128 - (37.945 * img1[i][j][0] / 256) - (74.494 * img1[i][j][1] / 256) + (112.439 * img1[i][j][2] / 256)
            elif k == 2:
                output[i][j][k] = 128 + (112.439 * img1[i][j][0] / 256) - (94.154 * img1[i][j][1] / 256) - (18.285 * img1[i][j][2] / 256)
                

plt.imshow(output)
plt.title("YCbCr Image")
plt.plot()

#print('Size of Original image is: ', img1.shape)
#print('Size of Output image is: ', output.shape)
```
