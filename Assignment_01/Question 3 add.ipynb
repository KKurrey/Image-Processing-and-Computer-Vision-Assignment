```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

image1 = cv2.imread('Input1.jpg')  
image2 = cv2.imread('Input2.jpg')

plt.savefig("Original1.jpg")
plt.savefig("Original2.jpg")

plt.imshow(image1)

plt.title("Input1 image")

plt.show()

plt.imshow(image2)

plt.title("Input2 image")

plt.show()

#Method1 - using CV2

t1 = time.time()

add = cv2.add(image1, image2)

cv2.imwrite('AddedImage1.jpg', add)

t2 = time.time()
time_cv2 = t2 - t1
print(time_cv2)

plt.imshow(add)

plt.title("Addition using CV2")

plt.show()

#Method2 - using def

t3 = time.time()

def addImage(img1, img2):
    return(cv2.add(image1, image2))

adding = addImage(image1, image2)

cv2.imwrite('AddedImage2.jpg', adding)

t4 = time.time()
time_def = t4 - t3
print(time_def)

plt.imshow(adding)

plt.title("Addition using def")

plt.show()

#Method3 - using list comprehension

t5 = time.time()

x, y, z = image1.shape

image3 = [image1[i] + image2[i] for i in range(x)]
image3 = np.asarray(image3, dtype = np.uint8)

t6 = time.time()
time_list = t6 - t5
print(time_list)

plt.imshow(image3)

plt.title("Addition using list Comprehension")

plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()

#Comparison of processing speed

x = ['Method-1', 'Method-2', 'Method-3']

Processing_times = [time_cv2, time_def, time_list]

x_pos = [i for i in range (len(x))]

plt.bar(x_pos, Processing_times, color = 'blue')
plt.xlabel("<-------------Method------------->  ")
plt.ylabel("<-------------Processing Time------------->  ")
plt.title(" Procesing Time versus method ")
plt.xticks(x_pos, x)

plt.show()
```
