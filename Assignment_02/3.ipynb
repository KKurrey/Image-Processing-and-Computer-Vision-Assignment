### Q. 3. Write python from scratch to convert RGB Image (lena.tif) to YUV color image using the following equation.

   Y =  0.299R + 0.587G + 0.114B

   U = -0.147R - 0.289G + 0.436B

   V =  0.615R - 0.515G - 0.100B

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.tif')
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#plt.imshow(img1)
#plt.title("Original Image")
#plt.plot()

output = img1

for k in range(img1.shape[2]):
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
            if k == 0:
                output[i][j][k] = 0.299 * img1[i][j][0] + 0.587 * img1[i][j][1] + 0.114 * img1[i][j][2]
            elif k == 1:
                output[i][j][k] = -0.147 * img1[i][j][0] - 0.289 * img1[i][j][1] + 0.436 * img1[i][j][2]
            elif k == 2:
                output[i][j][k] = 0.615 * img1[i][j][0] - 0.515 * img1[i][j][1] - 0.100 * img1[i][j][2]
                

plt.imshow(output)
plt.title("YUV Image")
plt.plot()

#print('Size of Original image is: ', img1.shape)
#print('Size of Output image is: ', output.shape)
```
